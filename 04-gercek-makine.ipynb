```python
from qiskit import * 
import matplotlib
```

```python
circuit = QuantumCircuit(2,2)
```

```python
circuit.draw()
```

```python
%matplotlíb inline
```

```python
circuit.draw(output='mpl')
```

```python
circuit.h(0)
circuit.draw(output='mpl')
```

```python
circuit.cx(0,1) #  0= kontrol qubit , 1 = target qubit
circuit.measure([0,1],[0,1])
circuit.draw(output='mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
```

```python
result = execute(circuit , backend= simulator).result()
```

```python
from qiskit.tools.visualization import plot_histogram

plot_histogram(result.get_counts(circuit))
```

```python
IBMQ.load_account()
provider = IBMQ.get_provider("ibm-q")
quantum_computer = provider.get_backend('ibmq_quito')
```

```python
job = execute(circuit,backend=quantum_computer)
```

```python
from qiskit.tools.monitor import job_monitor
job_monitor(job)
```

```python
quantum_result = job.result()
plot_histogram(quantum_result.get_counts(circuit))
```
